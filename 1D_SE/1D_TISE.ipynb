# Time-independent Schrödinger equation in 1D
For the time-dependent Schrödinger equation
$$
i \hbar \frac{\partial}{\partial t} \ket{\psi(t, x)} = \hat{H} \ket{\psi(t, x)}
$$
can be decomposed by the method of separation of variables when the potential is time-independent $V=V(x)$, that is
$$
\hat{H}\ket{\psi(x)} = E\ket{\psi(x)},
$$
where the hamiltonian is given by
$$
\hat{H} = -\frac{\hbar^2}{2 m} \nabla^2 + V(x).
$$

## Problem 

The goal of this project is to solve time-independent Schrödinger equation in 1-dimension called "Eigen Probelm", that is
$$
\boxed{\color{blue}
-\frac{\hbar^2}{2 m}\frac{d^2}{dx^2}\psi(x) + V(x)\psi(x) = E\psi(x),
}
$$
### Discrete Schrödinger equation
The time-independent Schrödinger equation is
$$
-\frac{\hbar^2}{2 m}\frac{d^2}{dx^2}\ket{\psi} + V\ket{\psi} = E\ket{\psi},
$$
where is the column vector
$$
\ket{\psi} = \psi_{i},\quad i=0,1,\ldots,n-1,
$$
potential $V$ is a diagonal matrix
$$
V = V_i\,\delta_{i,j},\quad i,j=0,1,\ldots,n-1,
$$
and the 1-dimension laplacian operator $\nabla^2$ approximate to a $n\times n$ matrix, since
$$
\frac{d^2f(x)}{dx^2} \approx \frac{f(x-\delta_x) - 2f(x) + f(x+\delta_x)}{\delta_x^2},
$$
we define
$$
\frac{d^2}{dx^2}f_i = \frac{1}{\delta_x} \left(f_{i-1}-2f_{i}+f_{i+1}\right),\quad i=0,1,\ldots,n-1,
$$
where $f_{-1}=f_n=0$, plugging we have
$$
-\frac{\hbar^2}{2 m}\frac{d^2}{dx^2}\psi_i + V_i\,\delta_{i,j}\psi_i = E\psi_i,
$$
or
$$
-\frac{\hbar^2}{2 m}\frac{1}{\delta_x^2}
\begin{pmatrix}
-2 & 1 & 0 & 0 & \cdots & 0 & 0 & 0 \\
1 & -2 & 1 & 0 & \cdots & 0 & 0 & 0 \\
0 & 1 & -2 & 1 & \cdots & 0 & 0 & 0 \\
0 & 0 & 1 & -2 & \cdots & 0 & 0 & 0 \\
\vdots & \vdots & \vdots & \vdots & \ddots & \vdots & \vdots & \vdots \\
0 & 0 & 0 & 0 & \cdots & -2 & 1 & 0 \\
0 & 0 & 0 & 0 & \cdots & 1 & -2 & 1 \\
0 & 0 & 0 & 0 & \cdots & 0 & 1 & -2
\end{pmatrix}
\begin{pmatrix}\psi_{0}\\\psi_{1}\\\psi_{2}\\\psi_{3}\\\vdots\\\psi_{n-3}\\\psi_{n-2}\\\psi_{n-1}\\\end{pmatrix}
+ 
\begin{pmatrix}
V_0 & 0 & 0 & 0 & \cdots & 0 & 0 & 0 \\
0 & V_1 & 0 & 0 & \cdots & 0 & 0 & 0 \\
0 & 0 & V_2 & 0 & \cdots & 0 & 0 & 0 \\
0 & 0 & 0 & V_3 & \cdots & 0 & 0 & 0 \\
\vdots & \vdots & \vdots & \vdots & \ddots & \vdots & \vdots & \vdots \\
0 & 0 & 0 & 0 & \cdots & V_{n-3} & 0 & 0 \\
0 & 0 & 0 & 0 & \cdots & 0 & V_{n-2} & 0 \\
0 & 0 & 0 & 0 & \cdots & 0 & 0 & V_{n-1}
\end{pmatrix}
\begin{pmatrix}\psi_{0}\\\psi_{1}\\\psi_{2}\\\psi_{3}\\\vdots\\\psi_{n-3}\\\psi_{n-2}\\\psi_{n-1}\\\end{pmatrix}
= E
\begin{pmatrix}\psi_{0}\\\psi_{1}\\\psi_{2}\\\psi_{3}\\\vdots\\\psi_{n-3}\\\psi_{n-2}\\\psi_{n-1}\\\end{pmatrix},
$$
So, now it becomes solving this eigen problem for a large system
$$
\begin{pmatrix}
\displaystyle \frac{\hbar^2}{\delta_x^2 m} +V_0&\displaystyle  -\frac{\hbar^2}{2\delta_x^2 m} & 0 & 0 & \cdots & 0 & 0 & 0 \\[2ex]
\displaystyle -\frac{\hbar^2}{2\delta_x^2 m} &\displaystyle  \frac{\hbar^2}{\delta_x^2 m} +V_1&\displaystyle  -\frac{\hbar^2}{2\delta_x^2 m} & 0 & \cdots & 0 & 0 & 0 \\[2ex]
0 &\displaystyle  -\frac{\hbar^2}{2\delta_x^2 m} &\displaystyle  \frac{\hbar^2}{\delta_x^2 m} +V_2&\displaystyle  -\frac{\hbar^2}{2\delta_x^2 m} & \cdots & 0 & 0 & 0 \\[2ex]
0 & 0 &\displaystyle  -\frac{\hbar^2}{2\delta_x^2 m} &\displaystyle  \frac{\hbar^2}{\delta_x^2 m} +V_3& \cdots & 0 & 0 & 0 \\[2ex]
\vdots & \vdots & \vdots & \vdots & \ddots & \vdots & \vdots & \vdots \\[2ex]
0 & 0 & 0 & 0 & \cdots &\displaystyle  \frac{\hbar^2}{\delta_x^2 m} +V_{n-3}&\displaystyle  -\frac{\hbar^2}{2\delta_x^2 m} & 0 \\[2ex]
0 & 0 & 0 & 0 & \cdots &\displaystyle  -\frac{\hbar^2}{2\delta_x^2 m} &\displaystyle  \frac{\hbar^2}{\delta_x^2 m} +V_{n-2}&\displaystyle  -\frac{\hbar^2}{2\delta_x^2 m} \\[2ex]
0 & 0 & 0 & 0 & \cdots & 0 &\displaystyle  -\frac{\hbar^2}{2\delta_x^2 m} &\displaystyle  \frac{\hbar^2}{\delta_x^2 m}+V_{n-1}
\end{pmatrix}
\begin{pmatrix}\psi_{0}\\[3.5ex]\psi_{1}\\[3.5ex]\psi_{2}\\[3.5ex]\psi_{3}\\[3.5ex]\vdots\\[3.5ex]\psi_{n-3}\\[3.5ex]\psi_{n-2}\\[3.5ex]\psi_{n-1}\\\end{pmatrix}
= E
\begin{pmatrix}\psi_{0}\\[3.5ex]\psi_{1}\\[3.5ex]\psi_{2}\\[3.5ex]\psi_{3}\\[3.5ex]\vdots\\[3.5ex]\psi_{n-3}\\[3.5ex]\psi_{n-2}\\[3.5ex]\psi_{n-1}\\\end{pmatrix}
$$

#### seting the position space
$$
x = [x_i,x_i+dx,\ldots,x_f]
$$

<div style="text-align:center"><img src="./assets/position_space.png" style="width:800px;margin:auto"></div>

````python
xi = 0      # initial point
xf = 1      # final point
dx = 0.001   # step (delta_x) 
x = np.arange(xi,xf,dx)
````

### define the potential function

````python
f=plt.figure()
def V(x):
    v = 1000 *(x-0.5)**2
    #v[x<0.5]=0
    return v
plt.plot(x,V(x))
plt.title("Potential $V(x)$")
plt.show()
f.savefig('assets/Potential.svg')
````

### Solving the eigen problem
Since the discrete Hamiltonian is an tridiagonal matrix in the form
$$\hat{H}=
\begin{pmatrix}
{b_{1}}&{c_{1}}&{}&{}&{0}\\
{a_{2}}&{b_{2}}&{c_{2}}&{}&{}\\
{}&{a_{3}}&{b_{3}}&\ddots &{}\\
{}&{}&\ddots &\ddots &{c_{n-1}}\\
{0}&{}&{}&{a_{n}}&{b_{n}}\\
\end{pmatrix},
$$
we can use [tridiagonal matrix algorithm](https://en.wikipedia.org/wiki/Tridiagonal_matrix_algorithm) to solve it, which only need  $\mathcal{O}(n)$ operations rather than $\mathcal{O}(n^3)$ operations by [Gaussian elimination](https://en.wikipedia.org/wiki/Gaussian_elimination).

Here using ```eigh_tridiagonal(d,e)``` function, built up by [scipy](https://docs.scipy.org/doc/scipy/reference/generated/scipy.linalg.eigh_tridiagonal.html).

````python
d = 1/dx**2 + V(x)
e = -1/(2*dx**2) * np.ones(len(x)-1)
w,v = eigh_tridiagonal(d,e)
````

### Define the Eigenfunction $\psi_{i}$ and Eigenvalue $E_i$
- Eigenfunction $\psi_{i}$, $i=0,1,\ldots,\text{length}\left(x\right)$
- Eigenvalue $E_i$, $i=0,1,\ldots,\text{length}\left(x\right)$

````python
phi = v.T
E = w
````

### Plot the Eigenfunction
$$
\left|\phi_{i}(x)\right|^2,\quad i =0,1,\ldots,9
$$

````python
row = 2
col = 5
f=plt.figure()
for i in range(row*col):
    plt.subplot(row,col,i+1)
    plt.plot(x,phi[i]**2)
    plt.title("$\phi_{"+str(i)+"}(x)$")
    plt.yticks([])
f.tight_layout()
f.savefig('assets/Eigenfunction.svg')
````

````python
E_N = 100
f, (a0,a1) = plt.subplots(1, 2, gridspec_kw={'width_ratios': [1, 10]})
for i in range(E_N): 
    a0.axhline(y = E[i])
a1.plot(E[0:E_N],'.',markersize=2)
a0.set_yticks([])
a0.set_xticks([])
a1.set_yticks([])
a1.set_xlabel('$n$')
a1.set_ylabel('Energy')
f.suptitle('Eigenvalue $E_n$')
f.savefig('assets/Eigenvalue.svg')
````
